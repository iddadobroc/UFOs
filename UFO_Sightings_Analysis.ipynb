```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("willianoliveiragibin/ufo-sightings")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Percorso al file del dataset
dataset_path = "/root/.cache/kagglehub/datasets/willianoliveiragibin/ufo-sightings/versions/1"

# Trova il file CSV nel percorso (se ci sono più file, prendi il primo)
file_name = [f for f in os.listdir(dataset_path) if f.endswith('.csv')][0]
file_path = os.path.join(dataset_path, file_name)

# Carica il file come DataFrame
df = pd.read_csv(file_path)

# Visualizza un'anteprima dei dati
print(df.head())
```

```python

df = df.drop(columns=["Unnamed: 0"])

df.head()

```

```python
# Controlla il numero di valori mancanti per ogni colonna
missing_values = df.isnull().sum()

# Mostra le colonne con valori mancanti e il numero di valori mancanti
print(missing_values[missing_values > 0])
```

```python
# Elimina le righe con missing values in tutte le colonne eccetto "Description"
df_cleaned = df.dropna(subset=["Country_Code", "Country", "Region", "Locale", "UFO_shape"])

# Controlla il nuovo DataFrame
print(df_cleaned.isnull().sum())  # Verifica i missing values rimasti
print(df_cleaned.shape)          # Verifica il numero di righe e colonne
```

```python
df_cleaned.head()
```

```python
# Valori unici nella colonna UFO_shape
unique_shapes = df["UFO_shape"].unique()

# Stampa i valori unici
print(unique_shapes)
```

```python
# Conta i valori nella colonna UFO_shape e ordina in ordine discendente
shape_counts = df["UFO_shape"].value_counts(ascending=False)

# Visualizza i risultati
print(shape_counts)
```

```python
# Sostituisci "Other" e "Unknown" con "Uncertain or unknown"
df["UFO_shape"] = df["UFO_shape"].replace({"Other": "Uncertain or unknown", "Unknown": "Uncertain or unknown"})

# Controlla le modifiche effettuate
print(df["UFO_shape"].value_counts())
```

```python
import matplotlib.pyplot as plt

# Controlla l'intervallo di anni nel dataset
print(f"Anni nel dataset: {df['Year'].min()} - {df['Year'].max()}")

# Crea un istogramma per visualizzare la distribuzione degli anni
plt.figure(figsize=(12, 6))
df["Year"].plot(kind="hist", bins=range(df["Year"].min(), df["Year"].max() + 1), edgecolor="black", color="skyblue")
plt.title("Distribution of UFO Sightings by Year", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Sightings", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crea un istogramma con bin per intervalli di 5 anni
bin_width = 5
bins = range(df["Year"].min(), df["Year"].max() + bin_width, bin_width)

plt.figure(figsize=(12, 6))
df["Year"].plot(kind="hist", bins=bins, edgecolor="black", color="skyblue")
plt.title("Distribution of UFO Sightings by Year (Grouped by 5 Years)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Sightings", fontsize=12)

# Aggiorna i tick dell'asse x per mostrarli chiaramente
plt.xticks(bins, rotation=45, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Conta la frequenza di avvistamenti per stagione
season_counts = df["Season"].value_counts()

# Crea il bar chart
plt.figure(figsize=(8, 5))
season_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("UFO Sightings by Season", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Number of Sightings", fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Conta la frequenza di avvistamenti per ogni ora
hour_counts = df["Hour"].value_counts().sort_index()  # Ordina per ordine cronologico

# Crea il bar chart
plt.figure(figsize=(12, 6))
hour_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("UFO Sightings by Hour of the Day", fontsize=16)
plt.xlabel("Hour (24-hour format)", fontsize=12)
plt.ylabel("Number of Sightings", fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Conta il numero di avvistamenti per paese
country_counts = df["Country"].value_counts()

# Filtra i paesi con almeno 30 avvistamenti
country_counts_filtered = country_counts[country_counts >= 30]

# Crea il bar chart
plt.figure(figsize=(12, 6))
country_counts_filtered.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("UFO Sightings by Country (Countries with ≥30 Sightings)", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Sightings", fontsize=12)

# Ruota le etichette dell'asse x per evitare sovrapposizioni
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostra il grafico
plt.tight_layout()
plt.show()
```
